# Pandas 기초  
Pandas에서는 데이터 타입을 크게 Series와 DataFrame으로 두가지로 나뉘어서 다루고 있다.  


## Series  
series 라는 것은 pandas의 1차원적인 데이터 타입을 말한다 

```python
import pandas as pd
import numpy as np

series = pd.Series([1, 3, 5, np.nan, 7, 9])
series
```

```python
# Series에서 slice 하면 true 값들만 반환
series[series > 3.0]
```

## DataFrame
우리가 Excel에서 보듯이 2차원적인 데이터를 저장된 형태.  
Series가 모이면 DataFrame이라고 생각하면 된다. 

```python
data = {'Name': 'DongSoo',
        'Age': 21,
        'Height': 180.2,
        'Weight': 58.5,
        'Job': 'student'}

columns = ['Name', 'Age', 'Height', 'Weight', 'Job']

df = pd.DataFrame(data=data, index=[0], columns=columns)  # index는 임의로 0으로 주었다. 
df
```

# Boolean Indexing

```python
df['alcohol'] > 14  # 한 Column에 조건을 줄 수 있다
```

```python
df[df['alcohol'] > 14]  # 위에서 준 조건을 True 값만 반환할 수 있다
```

```python
isin_row = df['alcohol'].isin([14.23, 14.20, 14.22])  # 원하는 value 값이 있는 것들만 True로 반환

df[isin_row]
```

# Process DataFrame

```python
# Nan 처리
df = df.dropna()  # Nan이 있는 Row들을 없애준다
df.fillna(value=0, inplace=True)  # value: Nan 값들을 0으로 채운다, inplace df 내부에 실행한다
```

```python
mean = df['alcohol'].mean()  # Nan으로 비워진 곳들을 mean으로 채울 수 있다
mean_filled = df.fillna(value=mean)
```

```python
renamed = df.rename(columns={'alcohol':'ALCOHOL'})  # Column 명 바꾸기
renamed.head()
```
